```python
# Q1 : save the artists, albums and tracks tables in three df  
```

```python
QUERY = "SELECT * from artists"
artists = pd.read_sql(QUERY, conn)
artists.head(1)
```

```python
QUERY = "SELECT * from albums"
albums = pd.read_sql(QUERY, conn)
albums.head(1)
```

```python
QUERY = "SELECT * FROM tracks"
tracks = pd.read_sql(QUERY, conn)
tracks.head(1)
```

```python
#Q2: Join artists and albums tables 
#    Show TWO different ways of getting this result 
```

```python
df1 = artists.merge(albums, left_on='ArtistId', right_on = 'ArtistId', how = 'inner')
df1.head()
```

```python
df2 = pd.read_sql_query("SELECT * FROM artists INNER JOIN albums ON albums.`ArtistId` = artists.`ArtistId`", conn)
df2.head()
```

```python
len(df1) == len(df2)
```

```python
#Q3: Create a PivotTable showing for each artist, his/her/their albums (the titles), and for each album the duration 
# 2 levels expected in index 
#Hint : Use the table tracks and the df created in Q2 
```

```python
df3 = df1.merge(tracks,left_on = 'AlbumId', right_on='AlbumId', how = 'inner')
df3.head()
```

```python
df4 = pd.pivot_table(df3, index = ['Name_x','Title'], values=['Milliseconds'], aggfunc=sum)
df4.head()
```

```python
#Q4: Plot a decreasing bar char showing the top 20 album durations 
```

```python
df4['Milliseconds'].sort_values(ascending = False).head(20).plot(kind = 'bar', figsize = (10,5))
```

```python
#Q5: Create an UNpivoted dataFrame with as many rows as there are invoice_items
# for each invoice_item (it means each InvoiceLineId), country of the customer and genre of the track are expected
# Use InvoiceLineId as Index
#Hint: make multiple merging operations using customers, invoices, invoice_items, tracks and genres (repetitive task)
#       Use the argument suffixes of the function pd.DataFrame.merge for overlapping column names
```

pd.DataFrame.merge documentation

http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html 


```python
QUERY = "SELECT * FROM customers"
customers = pd.read_sql(QUERY, conn)
customers.head(1)
```

```python
QUERY = "SELECT * FROM invoices"
invoices = pd.read_sql(QUERY, conn)
invoices.head(1)
```

```python
QUERY = "SELECT * FROM invoice_items"
invoice_items = pd.read_sql(QUERY, conn)
invoice_items.head(1)
```

```python
tracks.head(1)
```

```python
QUERY = "SELECT * FROM genres"
genres = pd.read_sql(QUERY, conn)
genres.head(1)
```

```python
df5 = customers.merge(invoices, left_on='CustomerId', right_on = 'CustomerId', how = 'inner')
df5 = df5.merge(invoice_items, left_on='InvoiceId', right_on = 'InvoiceId', how = 'inner')
df5 = df5.merge(tracks, left_on='TrackId',right_on='TrackId', how = 'inner')
df5 = df5.merge(genres, left_on='GenreId',right_on='GenreId', how = 'inner', suffixes = ['_track','_genre'])
```

```python
df5.set_index('InvoiceLineId', inplace = True)
```

```python
df5 = df5.loc[:,['Country', 'Name_genre']]
df5.head()
```
